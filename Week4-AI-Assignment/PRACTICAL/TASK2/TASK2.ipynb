```python
from transformers import pipeline

#Load GPT2 Models from hugging Face
generator = pipeline("text-generation", model="gpt2")
print("This is GPT2 Models powered by JsMummie Herself")

while True:
  prompt = input("You: ").strip()
  if prompt.lower() in {"exit", "quit"}:
    print("GoodBye!, Remember to code on daily basis")
    break

  #Generate the text/output
  output = generator(prompt, max_new_tokens=1000, temperature=0.7)
  print("\nAI: ", output[0]["generated_text"], "\n")
```
